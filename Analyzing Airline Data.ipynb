### Introduction

As a group, we share common love for air-travel and curiosity about the airline industry. The recent missing Malaysian airline plane related incident spurred our interest further and made us realize how little we know about airlines, their schedules and their function in general. We began researching and stumbled upon US DoT dataset on their website, which looked pretty exhaustive and promising.

The United States Department of Transportation tracks the on-time performance of domestic flights operated by large air carriers. Various airlines have been contributing to the US DoT by sharing their yearly flight data with the U.S. DoT since 1987. The U.S. Department of Transportation publishes a monthly summary of airline on-time performance, including causes of delay, in the Air Travel Consumer Report along with an annual reports on variety of on-time and flight delay data so that the public would have clear information about the nature and sources of airline delays and cancellations.

DoT maintains each year’s data in the form of CSV files. Each file contains on-time arrival data for non-stop domestic flights by major air carriers, and provides such additional items as departure and arrival delays, origin and destination airports, flight numbers, scheduled and actual departure and arrival times, cancelled or diverted flights, taxi-out and taxi-in times, air time, and non-stop distance.

By exploring this dataset, we wanted to understand delay trends in the US airline industry. We planned to find out the most important/popular carriers in the country, evaluate their performance - about delays and the causes for the delays, what cities they affect. Our intent was to unravel some interesting trends in the last few years in the US airline industry with the help of data visualizations.

## Analysis#1 - Delays by Airport

Here we analyze the delays by airports. It will be interesting to find which airports have the most delays and what is the avergae delay time.

```python
"""
Aggregating the Master dataframe by Origin and Destination airport to get the mean and count of delays for each airport.
Sorting the individual Arrival delay and Departure delay dataframes to get the top 20 and bottom 20 records.
Reseting the index to get the Airport as a column and renaming individual 'Destination' and 'Origin' airport column names to 'Ariport'
"""

dfArrvAirportStats = dfMaster[['ArrDelay','Dest_airport']][dfMaster['ArrDelay'] > INPUT_DELAY].groupby([dfMaster['Dest_airport']]).agg([ np.mean, np.std, len])
dfDepAirportStats = dfMaster[['Origin_airport','DepDelay']][dfMaster['DepDelay'] > INPUT_DELAY].groupby([dfMaster['Origin_airport']]).agg([np.mean, np.std, len])
#dfArrvAirportStats.index.set_names(["Aiport"], inplace=True)
#dfDepAirportStats.index.set_names(["Aiport"], inplace=True)
#dfAirportStats = pd.concat((dfArrvAirportStats , dfDepAirportStats), axis = 1)
dfArrvAirportStats = dfArrvAirportStats.sort([('ArrDelay','len')], ascending=[0])
dfDepAirportStats = dfDepAirportStats.sort([('DepDelay','len')], ascending=[0])
dfArrvAirportStats.reset_index(inplace=True)
dfDepAirportStats.reset_index(inplace=True)
dfArrvAirportStats = dfArrvAirportStats.rename(columns={'Dest_airport' : 'Airport'})
dfDepAirportStats = dfDepAirportStats.rename(columns={'Origin_airport' : 'Airport'})
#dfAirportStats.Airport = dfAirportStats.Airport.astype(str)
```

Here we want to analyze relationship between days of months and delays. It will be interesting to find if some days/weeks of a month have more delays. We use "DayOfMonth" across all months for this analysis.

```python
# Code to create a new aggregated dataframes dfDayofMonthDepDelayStats and dfDayofMonthDepArrStats from dfMaster. We then plot the graphs using these two dataframes.
# We plot count of delays and average delay for every day of month. We plot Arrival and Departure delay separately.

dfDayofMonthDepDelayStats = dfMaster[['DayofMonth','DepDelay']][dfMaster['DepDelay'] > 15].groupby([dfMaster['DayofMonth']]).agg([np.mean,len])

del dfDayofMonthDepDelayStats['DayofMonth']
dfDayofMonthDepDelayStats.reset_index(inplace=True)
dfDayofMonthDepDelayStats.columns

labels = xrange(1,32)
fig = plt.figure()
fig.suptitle('Departure Delays by Day of Month', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.95)
ax.set_xlabel('Day of Month')
ax.set_ylabel('Average Departure Delay(mins)')
ax.plot(dfDayofMonthDepDelayStats['DayofMonth'], dfDayofMonthDepDelayStats['DepDelay']['mean'])
plt.xticks(dfDayofMonthDepDelayStats['DayofMonth'],labels, rotation=45)
ax.lines[0].set_color(DEPARTURE_COLOR)

fig = plt.figure()
fig.suptitle('Departure Delays by Day of Month', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.95)
ax.set_xlabel('Day of Month')
ax.set_ylabel('No. of flights')
ax.plot(dfDayofMonthDepDelayStats['DayofMonth'], dfDayofMonthDepDelayStats['DepDelay']['len'])
plt.xticks(dfDayofMonthDepDelayStats['DayofMonth'],labels, rotation=45)
ax.lines[0].set_color(DEPARTURE_COLOR)


dfDayofMonthDepArrStats = dfMaster[['DayofMonth','ArrDelay']][dfMaster['ArrDelay'] > 15].groupby([dfMaster['DayofMonth']]).agg([np.mean,len])
del dfDayofMonthDepArrStats['DayofMonth']
dfDayofMonthDepArrStats.reset_index(inplace=True)
dfDayofMonthDepArrStats.columns


fig = plt.figure()
fig.suptitle('Arrival Delays by Day of Month', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.95)
ax.set_xlabel('Day of Month')
ax.set_ylabel('Average Arrival Delay(mins)')
ax.plot(dfDayofMonthDepArrStats['DayofMonth'], dfDayofMonthDepArrStats['ArrDelay']['mean'])
plt.xticks(dfDayofMonthDepDelayStats['DayofMonth'],labels, rotation=45)
ax.lines[0].set_color(ARRIVAL_COLOR)

fig = plt.figure()
fig.suptitle('Arrival Delays by Day of Month', fontsize=14, fontweight='bold')
ax = fig.add_subplot(111)
fig.subplots_adjust(top=0.95)
ax.set_xlabel('Day of Month')
ax.set_ylabel('No. of flights')
ax.plot(dfDayofMonthDepArrStats['DayofMonth'], dfDayofMonthDepArrStats['ArrDelay']['len'])
plt.xticks(dfDayofMonthDepArrStats['DayofMonth'],labels, rotation=45)
ax.lines[0].set_color(ARRIVAL_COLOR)
```

3. In the middle of the month, the average delay in minutes and total number of delayed flights is lower compared to initial and last weeks.
